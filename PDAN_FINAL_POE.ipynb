###  **Image Generators**

**Imports, Directories and Image Sizes**
<br>
<br>
We start off by using Tensorflow and Keras for our image processing. Tensorflow is a useful tool in image proceessing, specifically TensorFlow is a robust deep learning framework, and Keras is a high-level API that provides a modular, easy-to-use, and organised interface to solve real-life deep learning problems.

```python
# IMPORT THAT GENERATES BATCHES OF TENSOR IMAGE DATA WITH REAL-TIME DATA AUGMENTATION
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# DIRECTORIES
train_img_path = '/content/train' # TRAIN DIRECTORY
test_img_path = '/content/test' # TEST DIRECTORY

# IMAGE CONSTRAINTS OR SIZES

# WE SET WIDTH AND HEIGHT OF IMAGE TO 32
# SINCE WE'LL USE CNNs IN OUR ANALYSIS, WE NEED TO RESIZE THE IMAGES' HEIGHT/WIDTH BY A MULTIPLE OF THE FACTOR BY THE WHICH CNN REDUCES SPATIAL SIZE BY THE LAST LAYER
width,height = 32,32

# BATCH SIZE IS A NUMBER OF SAMPLES PROCESSED BEFORE THE MODEL IS UPDATED
batch_size = 32 # BATCH SIZE OF 32 GAVE US THE BEST RESULT
```

**Train Generator**
<br>
<br>
This method will call all the train images from the train directory, and pass it through an image generator for processing and generating realistic images out of them.

```python
# TRAIN GENERATOR METHOD
def train():

  # DECLARE A VARIABLE FOR THE TRAIN GENERATOR
  train_data_gen = ImageDataGenerator(rescale = 1.0/255) # WE DIVIDE AN IMAGE BY 255 TO GET A RANGE OF 0 TO 1
  # # FLOW TRAINING IMAGES IN 32 BATCHES USING TRAIN_DATA_GEN GENERATOR FROM THE TRAIN DIRECTORY
  train_Dataset = train_data_gen.flow_from_directory(
    train_img_path,
    target_size = (width,height),
    batch_size = batch_size,
    class_mode = 'binary',
    shuffle = True)

  return train_Dataset # RETURN TRAINING IMAGES
```

```python
# CREATE THE TRAIN GENERATOR OBJECT FOR CALLING
train = train()

# DISPLAY THE SIZE OF A TRAIN IMAGE
train.image_shape
```

**Test Generator**
<br>
<br>
This will call all the test images from the test directory, and pass it through an image generator for processing and generating realistic images out of them.

```python
# METHOD FOR TEST GENERATOR
def test():

  # DECLARE A VARIABLE FOR THE TEST GENERATOR
  test_data_gen = ImageDataGenerator(rescale = 1.0/255) # WE DIVIDE AN IMAGE BY 255 TO GET A RANGE OF 0 TO 1

  # # FLOW VALIDATION OR TEST IMAGES IN BATCHES OF 32 USING TEST_DATA_GEN GENERATOR
  test_Dataset = test_data_gen.flow_from_directory(
    test_img_path,
    target_size = (width,height),
    batch_size = batch_size,
    class_mode = 'binary',
    shuffle = False)

  return test_Dataset # RETURN TEST IMAGES
```

```python
# CREATE THE TEST GENERATOR OBJECT FOR CALLING
test = test()

# DISPLAY SIZE OF A TEST IMAGE
test.image_shape
```

###  **Convolutional Neural Network (CNN)**

**Building the CNN Model**
<br>
<br>
A CNN is useful in finding patterns in images to recognise objects, classes, and categories. Therefore, we can say a CNN is our discriminator that takes an image and its respective target image and gives probability value for its similarity.

```python
# IMPORTS FOR CREATING THE CNN MODEL
from tensorflow.keras import layers, models

# METHOD FOR CNN MODEK
def cnn():
  # DEFINE THE MODEL OBJECT
  model = models.Sequential()

  # CREATE CONVOLUTIONAL LAYERS
  # WE USE THESE LAYERS TO BLUR AND SHARPEN IMAGES, BUT ALSO TO PERFORM OTHER OPERATIONS
  model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(width, height, 3)))
  model.add(layers.MaxPooling2D((2, 2)))
  model.add(layers.Conv2D(64, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D((2, 2)))
  model.add(layers.Conv2D(128, (3, 3), activation='relu'))
  model.add(layers.MaxPooling2D((2, 2)))

  # FLATTEN LAYER
  # WE DO THIS TO MAKE THE MULTIDIMENSIONAL INPUT ONE-DIMENSIONAL
  model.add(layers.Flatten())

  # DENSE LAYERS
  # THESE CAN CLASSIFY IMAGES BASED ON OUTPUT FROM CONVOLUTIONAL LAYERS
  model.add(layers.Dense(256, activation='relu'))
  model.add(layers.Dropout(0.5))  # OPTIMAL DROPOUT FOR REGULARISATION
  model.add(layers.Dense(1, activation='sigmoid'))  # OUTPUT LAYER WITH ONE NEURON (BINARY CLASSIFICATION)

  # COMPILE THE MODEL
  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

  # DISPLAY A SUMMARY OF THE MODEL ARCHITECTURE
  model.summary()

  return model # RETURN MODEL
```

```python
# CREATE THE CNN MODEL OBJECT
CNN = cnn()
```

**Training the Model**
<br>
<br>
Now we fit the CNN Model to the image data

```python
# DECLARE A VARIABLE FOR FITTING THE MODEL
# THE NUMBER OF EPOCHS DETERMINES HOW A LEARNING ALGORITHM WILL WORK IN THE TRAINING DATASET
# VALIDATION SPLIT SPECIFIES HOW MUCH OF THE TRAINING DATA WILL BE USED
history = CNN.fit(train,
                    epochs=10,
                    validation_data=test)
```

###  **Model Evaluation**

**Pipeline**
<br>
<br>
A pipeline performs multiple operations in a sequence, for each element of an iterable, in such a way that the output of each element is the input of the next.

```python
# IMPORT THE PIPELINE LIBRARY
from sklearn.pipeline import Pipeline

# DECLARE THE PIPELINE OBJECT BY PASSING THROUGH THE CNN MODEL
pipe = Pipeline(steps=[('CNN', cnn())])
```

```python
# FIT THE PIPELINE WITH TRAIN METHOD
pipe.fit(train)
```

```python
# FIT THE PIPELINE WITH TEST METHOD
pipe.fit(test)
```

**Predictions**
<br>
<br>
Here we test the accuracy of predictions

```python
# DECLARE A VARIABLE FOR PREDICTIONS
predictions = pipe.predict(test)
```

```python
# THEN CREATE A VARIABLE FOR TESTING THE PREDICTIONS
binary_predictions = (predictions > 0.5).astype(int)

# DISPLAY THE TEST PREDICTIONS
binary_predictions
```
